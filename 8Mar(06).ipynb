```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Define the parameter lambda for the Poisson distribution
lam = 3

# Create an array of x values for which to calculate the CDF
x_values = np.arange(0, 16)

# Calculate the CDF of the Poisson distribution for each x value
cdf = poisson.cdf(x_values, lam)

# Plot the CDF of the Poisson distribution
plt.plot(x_values, cdf, 'bo-', label='Poisson CDF')

# Add labels and legend to the plot
plt.xlabel('x')
plt.ylabel('Cumulative Probability')
plt.title('Cumulative Distribution Function of Poisson Distribution')
plt.legend(loc='best')

# Show the plot
plt.show()
```
